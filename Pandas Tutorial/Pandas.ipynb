# Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')

# Find average HP or Attack grouped by Type  
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)

# Find sum of Total grouped by Type  
df.groupby(['Type 1']).sum().sort_values('Total', ascending=False)

# Count Type 1 and Type 2 
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

# Large Datasets

```python
new_df = pd.DataFrame(columns=df.columns)

# Import 5 rows at a time
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    
new_df
```
